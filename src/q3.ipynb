# Functions to find and decompose the Projection Matrix

```python
def get_projection_matrix(world_coord, proj_coord):
    M = []
    for i in range(world_coord.shape[0]):
        M.append([
            -world_coord[i][0], -world_coord[i][1],
            -world_coord[i][2], -1, 0, 0, 0, 0,
            proj_coord[i][0]*world_coord[i][0],
            proj_coord[i][0]*world_coord[i][1],
            proj_coord[i][0]*world_coord[i][2],
            proj_coord[i][0],
        ])
        M.append([
            0, 0, 0, 0,
            -world_coord[i][0], -world_coord[i][1],
            -world_coord[i][2], -1,
            proj_coord[i][1]*world_coord[i][0],
            proj_coord[i][1]*world_coord[i][1],
            proj_coord[i][1]*world_coord[i][2],
            proj_coord[i][1],
        ])
    M = np.array(M)
    
    u, s, v = np.linalg.svd(M)
    p = v[-1, :]
    
    P = np.reshape(p, (3, 4))

    return P

def get_params(P):
    H = P[:3, :3]
    h = P[:, 3]
    Hinv = np.linalg.inv(H)

    Xo = np.matmul(-Hinv, h)
    q, r = np.linalg.qr(Hinv)
    r = np.linalg.inv(r)
    r /= r[2, 2]
    r[:, 0] = -r[:, 0]
    q = q.T
        
    return P, Xo, q, r
```

# Zhang's Method

```python
def get_rot_mat(r):
    theta = r[0];
    rot_x = np.array([[1, 0, 0],
                      [0, np.cos(theta), -np.sin(theta)],
                      [0, np.sin(theta), np.cos(theta)]])
    theta = r[1]
    rot_y = np.array([[np.cos(theta), 0, np.sin(theta)],
                      [0, 1, 0],
                      [-np.sin(theta), 0, np.cos(theta)]])
    theta = r[2]
    rot_z = np.array([[np.cos(theta), -np.sin(theta), 0],
                      [np.sin(theta), np.cos(theta), 0], 
                      [0, 0, 1]])
    return np.dot(rot_z, np.dot(rot_x, rot_y))



x, y = np.meshgrid(range(8), range(6))
world_coords = np.hstack((x.reshape(48, 1), y.reshape(48, 1), np.zeros((48, 1)))).astype(np.float32)



world_points = []
proj_points = []

for q in range(0, 8):
    img = cv2.imread('../Camera_calibration_data/myCam/z'+str(q)+'.jpg')
    
    ret, corners = cv2.findChessboardCorners(img, (8, 6))
    
    world_points.append(world_coords)
    proj_points.append(corners)

ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(world_points, proj_points, (img.shape[1],img.shape[0]), None, None)

print("Camera Calibration Matrix ->")
with np.printoptions(suppress=True):
    print(mtx)
    
    
    
    
x, y = np.meshgrid(range(8), range(6))
world_coords = np.hstack((x.reshape(48, 1), y.reshape(48, 1), np.zeros((48, 1)), np.ones((48, 1)))).astype(np.float32)
for q in range(0, 5):
    img = cv2.imread('../Camera_calibration_data/myCam/z'+str(q)+'.jpg')

    R = get_rot_mat(rvecs[q])
    RRC = np.hstack((R, tvecs[q][:]))
    P = np.dot(mtx, RRC)
    P /= P[-1, -1]

    proj_coords = []
    for i in range(48):
        projection = np.dot(P, world_coords[i])
        proj_coords.append([projection[0]/projection[2], projection[1]/projection[2]])
    proj_coords = np.array(proj_coords)

    for i in range(proj_coords.shape[0]):
        if i-8 >= 0:
            plt.plot([proj_coords[i][0], proj_coords[i-8][0]], [proj_coords[i][1], proj_coords[i-8][1]], 'ro-')
        if i-1 >= 0 and i%8 != 0:
            plt.plot([proj_coords[i][0], proj_coords[i-1][0]], [proj_coords[i][1], proj_coords[i-1][1]], 'ro-')
        if i+8 < proj_coords.shape[0]:
            plt.plot([proj_coords[i][0], proj_coords[i+8][0]], [proj_coords[i][1], proj_coords[i+8][1]], 'ro-')
        if i+1 < proj_coords.shape[0] and (i+1)%8 != 0:
            plt.plot([proj_coords[i][0], proj_coords[i+1][0]], [proj_coords[i][1], proj_coords[i+1][1]], 'ro-')


    plt.imshow(img)
    plt.show()
```
